```python
import os
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from helpers import load_data, standardize


data_directory = '../data'
train_dataset_path = os.path.join(data_directory, 'train.csv')

ids, labels, feature_names, raw_features = load_data(train_dataset_path)
print('Successfully loaded the data')
```

```python
# Get the PRI tau, lep and met features, which are 9 (3 of each)
# Index of the first feature to analize is 13 and of the last one is 21
PRItau_lep_met_features_index_start = 13
PRItau_lep_met_features_index_finish = 22 
numOfEach = 3

PRItau_lep_met_features_names = feature_names[PRItau_lep_met_features_index_start : PRItau_lep_met_features_index_finish]
PRItau_lep_met_features = raw_features[:, PRItau_lep_met_features_index_start : PRItau_lep_met_features_index_finish]

print("All together")
print(PRItau_lep_met_features_names)
print(PRItau_lep_met_features.shape)
#print(PRItau_lep_met_features)

# Just tau
PRItau_features_names = feature_names[PRItau_lep_met_features_index_start : PRItau_lep_met_features_index_start+numOfEach]
PRItau_features = raw_features[:,PRItau_lep_met_features_index_start : PRItau_lep_met_features_index_start+numOfEach]
print("Just tau")
print(PRItau_features_names)
print(PRItau_features.shape)
#print(PRItau_features)

# Just lep
PRIlep_features_names = feature_names[PRItau_lep_met_features_index_start+numOfEach : PRItau_lep_met_features_index_start+numOfEach*2]
PRIlep_features = raw_features[:,PRItau_lep_met_features_index_start+numOfEach : PRItau_lep_met_features_index_start+numOfEach*2]
print("Just lep")
print(PRIlep_features_names)
print(PRIlep_features.shape)
#print(PRIlep_features)

# Just met
PRImet_features_names = feature_names[PRItau_lep_met_features_index_start+numOfEach*2 : PRItau_lep_met_features_index_finish]
PRImet_features = raw_features[:,PRItau_lep_met_features_index_start+numOfEach*2 : PRItau_lep_met_features_index_finish]
print("Just met")
print(PRImet_features_names)
print(PRImet_features.shape)
#print(PRImet_features)
```

```python
# First we analyze tau
min_val_tau = np.min(PRItau_features, axis=0)
max_val_tau = np.max(PRItau_features, axis=0)
bins = np.linspace(min_val_tau, max_val_tau, 50)
plt.figure('Tau Usage')
ax = sns.histplot(x=PRItau_lep_met_features[:, 0], shrink=.8, multiple="dodge", hue=labels, discrete=True)
ax.set(title='Bar chart of the number of tau used in each experiment', xlabel='The number of jets', ylabel='The amount of experiments')
plt.show()

```
